```python

print("\nSoal 2")
# Total jumlah produksi sampah dari seluruh Kabupaten/Kota pada tahun tertentu
filtered_df_2018 = df_excel[df_excel['tahun'] == 2018]
tpy = {} #Dictinory untuk menyimpan nilai total produksi SAMPAH per tahun
# Iterasi dengan menggunakan .iterrows()
for i, baris in filtered_df_2018.iterrows():
    tahun_produksi = baris['tahun']
    total = baris['jumlah_produksi_sampah']
    tpy += total
print(f"Total produksi sampah pada seluruh Kabupaten/Kota pada tahun {tahun_produksi} adalah: {tpy:.2f}")
```

```python
print("\nSoal 3")
tpy = {}
# Total jumlah produksi sampah dari seluruh Kabupaten/Kota pertahun
for peryear in df_excel['tahun'].unique():
    filtered_df = df_excel[df_excel['tahun'] == peryear]
    tjp = filtered_df['jumlah_produksi_sampah']
    tjp = tjp.sum()
    tpy[peryear] = tjp
for peryear, total in tpy.items():
    print(f"Total produksi sampah pada seluruh Kabupaten/Kota pertahun {peryear} adalah: {total:.2f}")
```

```python
import pandas as pd

df_excel = pd.read_excel('data_set/jml_produksi_sampah.xlsx', sheet_name='data')
# Initialize a dictionary to store the sum for each year
sums_by_year = {}

# Use iterrows to iterate through the DataFrame and calculate the sums
for index, row in df_excel.iterrows():
    year = row['tahun']
    jumlah_produksi_sampah = row['jumlah_produksi_sampah']
    
    if year in sums_by_year:
        sums_by_year[year] += jumlah_produksi_sampah
    else:
        sums_by_year[year] = jumlah_produksi_sampah
    print(f"Print update data index ke-{index}: {sums_by_year};")

# # Print the sums for each year
# for year, sum_value in sums_by_year.items():
#     print(f"The sum of the 'jumlah_produksi_sampah' column for the year {year} is: {sum_value:.2f}")
```
