```python
print("Soal 1")
df_excel = pd.read_excel('dataset/jml_produksi_sampah.xlsx', sheet_name='data')
df_excel
```

```python

print("\nSoal 2")
filtered_df_2018 = df_excel[df_excel['tahun'] == 2018]
tpy = {} #Dictinory untuk menyimpan nilai total produksi SAMPAH per tahun

# Iterasi dengan menggunakan .iterrows()
for i, baris in filtered_df_2018.iterrows():
    tahun_produksi = baris['tahun']
    total = baris['jumlah_produksi_sampah'] * 365 # Ket: 365 = jumlah hari dalam setahun
    if tahun_produksi in tpy:
        tpy[tahun_produksi] += total
    else:
        tpy[tahun_produksi] = total

# Iterasi untuk mencetak jumlah produksi SAMPAH di tahun 2018
print(f"Total produksi sampah pada seluruh Kabupaten/Kota pada tahun {tahun_produksi} adalah: {tpy[2018]:.2f}")
```

```python
print("\nSoal 3")
tpy = {} # Dictinory untuk menyimpan nilai total produksi SAMPAH per tahun

# Iterasi dengan menggunakan .iterrows()
for i, baris in df_excel.iterrows():
    tahun_produksi = baris['tahun']
    jumlah_produksi = baris['jumlah_produksi_sampah'] * 365 # Ket: 365 = jumlah hari dalam setahun
    
    if tahun_produksi in tpy:
        tpy[tahun_produksi] += jumlah_produksi
    else:
        tpy[tahun_produksi] = jumlah_produksi

# Iterasi untuk mencetak jumlah produksi SAMPAH per tahun
for tahun, total in tpy.items():
    print(f"Total produksi sampah pada seluruh Kabupaten/Kota pada tahun {tahun} adalah: {total:.2f}")
```

```python
print("\nSoal 4")
tpkkt = {} # Total produksi tiap kota/kabupaten per tahun

# Iterasi dengan menggunakan .iterrows()
for i, baris in df_excel.iterrows():
    tahun_produksi = baris['tahun']
    kab_kot = baris['nama_kabupaten_kota']
    jml_pkk = baris['jumlah_produksi_sampah'] * 365 # Ket: 365 = jumlah hari dalam setahun
    
    if tahun_produksi not in tpkkt:
        tpkkt[tahun_produksi] = {}
    if kab_kot in tpkkt[tahun_produksi]:
        tpkkt[tahun_produksi][kab_kot] += jml_pkk
    else:
        tpkkt[tahun_produksi][kab_kot] = jml_pkk

dataset = []

# Iterasi untuk mencetak jumlah produksi SAMPAH tiap kota/kabupaten per tahun
for tahun, kab_kot_dic in tpkkt.items():
    for kab_kot, total in kab_kot_dic.items():
        # print(f"Total produksi sampah di {kab_kot} pada tahun {tahun} adalah: {total:.2f}")
        dataset.append({'Tahun Produksi': tahun, 'Kota/Kabupaten': kab_kot, 'Total Produksi': total})

# Ubah dataset menjadi dataframe
df = pd.DataFrame(dataset)
df # Hore ga pake print lagi (:
```

```python
print("\nExport DataFrame soal 4 ke excel/csv")
df.to_excel('output/hasil_df_soal4.xlsx', index=False)
df.to_csv('output/hasil_df_soal4.csv', index=False)
```
